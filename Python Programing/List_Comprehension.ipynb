```python
# List comprehension with range function

dummy_list=[i for i in range(11)]
print(dummy_list)
```

```python
dummy_list=[i ** i for i in range(1,11)]
print(dummy_list)
```

```python
# Creating a new list based on some operation on existing list

data=[15,36,78,26]
new_data=[i + 2 for i in data]
print(new_data)
```

```python
data=[15,36,78,26]
new_data=[i for i in data if i>=50]
print(new_data)
```

```python
data=[15,36,78,26]
new_data=[i for i in data if i%2==0]
print(new_data)
```

```python
data=[15,36,78,26]
new_data=[i+2 for i in data if i>=50]
print(new_data)
```

```python
data=[15,36,78,26]
new_data=[i+2 if i>=50 else i-2 for i in data]
print(new_data)
```

```python
data=[15,36,78,26]
data1=[1,2,3,4]
new_data=[{i:j} for i in data for j in data1]
print(new_data)
```
